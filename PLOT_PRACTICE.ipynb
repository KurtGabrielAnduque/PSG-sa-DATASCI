```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



df1 = pd.read_csv("/content/disease_fhsis_totals (1).csv")
df2 = pd.read_csv("/content/location.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
clone = df2.copy()
clone.drop(columns = ["Unnamed: 0","adm1_en","adm1_pcode","adm2_en","adm2_pcode","adm4_en","adm4_pcode","brgy_total_area"],inplace = True)
clone2 = df1.copy()
```

<font size = 5><b>MANIPULATING DATAS</b></font>

```python
merge_data = clone2.merge(clone, on = ["adm3_pcode"], suffixes = ('_left','_right'))
merge_data.head()
```

```python
merge_data["Year"] = merge_data["date"].str.split("-").str[0]
```

```python
#Top 5 Diseases by Total Cases (Bar Chart)
#Task: Create a bar chart showing the top 5 diseases with the highest total cases.

bar1 = merge_data.groupby(["disease_common_name"])["case_total"].sum().sort_values(ascending = False).head(5)
bar1.plot(figsize = (12,8), kind = "bar", color = "blue")
plt.title("Top 5 Highest Diseases")
plt.xlabel("DIFFERENT KINDS OF DESEASES")
plt.ylabel("Total Cases")
plt.xticks(rotation = 45)
plt.show()
```

```python
# Monthly Trend of Cases (Line Graph)
# Task: Extract the month from the date column and plot a line graph showing the trend of total cases per month.
merge_data["date"] = pd.to_datetime(merge_data['date'])
merge_data["Months"] = merge_data["date"].dt.month_name()
monthly_report = merge_data.groupby(["Months"])["case_total"].sum()

monthly_report.plot(figsize = (15,8), marker = "o", linestyle = "-", color = "salmon")
plt.title("monthly Trend of Cases")
plt.xlabel("Every month")
plt.ylabel("Total Cases")
plt.show()

```

```python
# Mortality Rate Comparison (Bar Chart)
'''
X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys')

plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
plt.legend()
plt.show()
'''
# Task: Create a side-by-side bar chart comparing cases vs deaths for the top 5 deadliest diseases.

bar3 = merge_data.groupby(["disease_common_name"])[["case_total","death_total"]].sum()
bar3 = bar3.sort_values(by = ["death_total"], ascending = False).head(5)

diseases = bar3.index
y_case = bar3["case_total"]
z_death = bar3["death_total"]

X_axis = np.arange(len(diseases))
plt.bar(X_axis - 0.2,y_case,0.4,label = 'Total Cases')
plt.bar(X_axis + 0.2,z_death,0.4,label = "Total Deaths")
plt.title("TOP 5 Deadliest Diseases")
plt.ylabel("Total Cases and Deaths  ")
plt.xlabel("Diseases")
plt.legend()
plt.xticks(X_axis, labels = diseases, rotation = 90 )
plt.show()
```

```python
# Case Distribution Across Cities (Horizontal Bar Chart)

# Task: Find the top 10 cities with the highest total cases and visualize them in a horizontal bar chart.

bar4 = merge_data.groupby(["adm3_en"])["case_total"].sum().sort_values(ascending = False).head(5)

plt.figure(figsize = (12,8))
plt.barh(bar4.index,bar4.values)
plt.gca().invert_yaxis()
plt.title("TOP 5 CITIES WITH THE HIGHEST CASES OF DISEASES")
plt.xlabel("Number of Cases")
plt.ylabel("City")
plt.show()
```
