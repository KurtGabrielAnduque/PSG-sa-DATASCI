```python
import pandas as pd

taxis = pd.read_csv("2019_Yellow_Taxi_Trip_Data.csv")
```

```python
taxis
```

```python
mask = taxis.columns.str.contains('id$|store_and_fwd_flag', regex = True)
columns_to_drop = taxis.columns[mask]
columns_to_drop
# here we save all the columns that has name that contains id or store_and_fwd_flag and we store it to columns_to_drop
```

```python
taxis = taxis.drop(columns = columns_to_drop) # here we remove the columns that we save from the columns_to_drop using the drop function
```

```python
taxis
```

```python
# renaming columns
# inorder for us to rename column we can use the function 'rename' inside the parethesis we will put the name of our columns in a dictionary type 
taxis = taxis.rename(
    columns = {
        'tpep_pickup_datetime' : 'pickup',
        'tpep_dropoff_datetime' : 'dropoff'
    }
)

taxis.columns
```

```python
taxis
```

```python
taxis[['pickup','dropoff']] = taxis[['pickup','dropoff']].apply(pd.to_datetime)
# here we apply a datatype to our columns pickup and dropoff with data type datetime
```

```python
taxis.dtypes # check for the updated data type of our dataframe
```

```python
taxis
```

```python
taxis['elapsed_time'] = taxis['dropoff'] - taxis['pickup']
```

```python
taxis = taxis.assign(
    elapsed_time = lambda x: x.dropoff - x.pickup,
    cost_before_tip = lambda x: x.total_amount - x.tip_amount,
    tip_pct = lambda x: x.tip_amount - x.cost_before_tip,
    fees = lambda x: x.cost_before_tip - x.cost_before_tip,
    avg_speed = lambda x: x.trip_distance.div(x.elapsed_time.dt.total_seconds()/60/60)
)
```

```python
taxis.sort_values(['passenger_count','pickup'],ascending = [False,True]).head()
# 
```

```python
taxis.sort_values(['fare_amount','tip_amount'], ascending = [False, True]).head()
```

```python
taxis.sort_values(['fare_amount']).head() #sort the dataframe based on the fare amount smallest to largest
```

```python
taxis.sort_values(['fare_amount','tip_amount'], ascending = [True, False]).head()
# when we are sorting values 
```

```python
taxis.nlargest(3,'elapsed_time') #the nlargest command takes the number of row of the largest value base sa  column na desired natin in this case the row with the largest elapsed_time
```

```python
taxis.nlargest(3,'trip_distance') # lets try it with largest trip distance
```
